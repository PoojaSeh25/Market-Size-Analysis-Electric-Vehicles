EV adoption over time 

```python
plt.figure(figsize=(12,6))
adoption_by_year=df['Model Year'].value_counts().sort_index()
sns.barplot(x=adoption_by_year.index, y=adoption_by_year.values, palette='viridis')
plt.title('EV Adoption Over Time')
plt.xlabel('Model Year')
plt.ylabel('Number of Vehicles Registered')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Distribution of EVs within the cities of those counties

```python
# Counting EV Distribution at the County Level
county_distibution= df['County'].value_counts()

# Selecting the Top Counties
top_counties=county_distibution.head(3).index

# Filtering the Dataset for the Top Counties
top_counties_data=df[df['County'].isin(top_counties)]

# Analyzing Distribution Within Cities of Top Counties
city_distribution_top_counties = top_counties_data.groupby(['County','City']).size().sort_values(ascending=False).reset_index(name='Number of Vehicles')

# Selecting the Top 10 Cities Across Counties
top_cities=city_distribution_top_counties.head(10)

# Visualizing  
plt.figure(figsize=(12, 8))
sns.barplot(x='Number of Vehicles', y='City', hue='County', data=top_cities, palette="magma")
plt.title('Top Cities in Top Counties by EV Registrations')
plt.xlabel('Number of Vehicles Registered')
plt.ylabel('City')
plt.legend(title='County')
plt.tight_layout()
plt.show()
```

Estimated Market Size Analysis of Electric Vehicles in the United States

```python
# EV registered every year
ev_registration_counts=df['Model Year'].value_counts().sort_index()
ev_registration_counts
```

```python
plt.figure(figsize=(12, 6))
plt.plot(ev_registration_counts.index, ev_registration_counts.values, marker='o', linestyle='-', color='brown')
plt.title("EVs Registered Every Year")
plt.xlabel("Model Year", fontsize=14)
plt.ylabel("Number of EVs Registered")
plt.tight_layout()
plt.show()
```

```python
from scipy.optimize import curve_fit

# Filter for years up to 2023
filtered_years = ev_registration_counts[ev_registration_counts.index <= 2023]

# Define the logistic growth model
def logistic_growth(x, K, r, t0):
    return K / (1 + np.exp(-r * (x - t0)))

# Prepare the data
x_data = np.array(filtered_years.index)  # Convert the index to a NumPy array
y_data = filtered_years.values

# Fit the logistic model to the data
initial_guess = [y_data.max(), 0.1, x_data.mean()]  # Initial parameters for curve fitting

# Increase maxfev to allow more iterations
params, covariance = curve_fit(logistic_growth, x_data, y_data, p0=initial_guess, maxfev=10000)

# Extract parameters
K, r, t0 = params

# Forecast the next five years
future_years = np.arange(2024, 2029)
forecasted_values = logistic_growth(future_years, K, r, t0)

# Create a DataFrame for better visualization of the forecast
forecast_df = pd.DataFrame({
    'Year': future_years,
    'Forecasted EV Registrations': forecasted_values
})

print(forecast_df)
```

```python
# Plot the historical data and forecasted values
plt.figure(figsize=(12, 6))

# Plot the historical data
plt.scatter(filtered_years.index, y_data, color='brown', label='Historical EV Registrations', zorder=5)

# Plot the logistic model fit for historical data
x_range = np.linspace(x_data.min(), x_data.max(), 1000)
plt.plot(x_range, logistic_growth(x_range, *params), color='pink', label='Logistic Growth Fit', linewidth=2)

# Plot the forecasted data
plt.plot(future_years, forecasted_values, color='green', label='Forecasted EV Registrations', linestyle='--', linewidth=2)

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Number of EV Registrations')
plt.title('Estimated Market Size: EV Registrations with Logistic Growth Model')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
